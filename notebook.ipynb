```python
# Use this cell to begin your analysis, and add as many as you would like!
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

csv = pd.read_csv('datasets/office_episodes.csv')
csv.head()
```

```python
# Defining variables
episode_number = csv['episode_number']
viewership_mil = csv['viewership_mil']
scaled_ratings = csv['scaled_ratings']
has_guests = csv['has_guests']
guest_stars = csv['guest_stars']
```

```python
# Conditions for colour scale and marker sizes
col = np.where(scaled_ratings < 0.25, 'red', 
               np.where(scaled_ratings < 0.5, 'orange', 
                        np.where(scaled_ratings < 0.75, 'lightgreen', 'darkgreen')))
size = np.where(has_guests==False,250,25)

# Create plot
fig, ax = plt.subplots()
ax.scatter(episode_number, viewership_mil, c=col, s=size)
ax.set_xlabel("Episode Number")
ax.set_ylabel("Viewership (Millions)")
ax.set_title("Popularity, Quality, and Guest Appearances on the Office")
plt.rcParams['figure.figsize'] = [11, 7]
ax.grid(True)
fig.tight_layout()

# Show plot
plt.show()
```
